## Bubble Graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['ChatGPT-4.1', 'ChatGPT-4o', 'LLaMA-4 Scout', 'LLaMA-4 Maverick', 'Gemini 2.5 Flash Lite', 'Gemini 2.5 Flash']
parameters = [1760, 1760, 109, 400, 20, 100]  # in billions
mmmu_scores = [68.4, 62.3, 58.2, 65.1, 55.0, 61.2]  # in %
training_compute = [1.5e8, 1.2e8, 2.5e7, 5.0e7, 5.0e6, 2.0e7]  # in PF-s

# Scale bubble sizes (arbitrary scaling for visualization)
bubble_sizes = [p * 10 for p in parameters]  # Scale parameters for bubble size

# Create bubble plot
plt.figure(figsize=(10, 6))

# Scatter plot with bubble sizes and color gradient
scatter = plt.scatter(parameters, mmmu_scores, s=bubble_sizes, c=training_compute, cmap='viridis', alpha=0.6, edgecolors='w', linewidth=0.5)

# Add labels for each point
for i, model in enumerate(models):
    plt.text(parameters[i], mmmu_scores[i], model, fontsize=9, ha='center', va='center')

# Set log scale for x-axis (parameters range from 20B to 1760B)
plt.xscale('log')
plt.xlabel('Parameters (Billions, Log Scale)')
plt.ylabel('MMMU Score (%)')

# Add colorbar for training compute
cbar = plt.colorbar(scatter)
cbar.set_label('Training Compute (PF-s)')

# Set title
plt.title('Vision Language Models Parameters')

# Adjust layout to prevent label clipping
plt.tight_layout()

# Show plot
plt.show()
```
