# Generate Bar Graph

```python
# Original vs Second across 6 models
import numpy as np
import matplotlib.pyplot as plt

# ---- EDIT THESE VALUES ----
models = ["ChatGPT-4.1","ChatGPT-4o", "Flash 2.5","Flash-Lite 2.5", "LLaMA-4 Maverick","LLaMA-4 Scout"]
original = [136, 198, 330, 301, 201, 66]  # Excel Stsats
second   = [305, 241, 97, 16, 227, 254]   # Excel Statss
# ---------------------------

if len(models) != 6:
    raise ValueError("Please provide exactly 6 model names.")
if not (len(original) == len(second) == len(models)):
    raise ValueError("original and second must each have exactly 6 numbers.")

x = np.arange(len(models))
width = 0.38

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, original, width, label="Original Prompt")
bars2 = ax.bar(x + width/2, second, width, label="EPP")

ax.set_title("Analysis of Original Prompt vs Error-Probabing Prompt (EPP)")
ax.set_ylabel("Weed Detected")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Annotate bars with values
for bars in (bars1, bars2):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f"{height:g}",
                    xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha="center", va="bottom")

plt.tight_layout()
# Optional: save an image copy near the notebook
plt.savefig("original_vs_second.png", dpi=200, bbox_inches="tight")
plt.show()
```

```python
# Original vs Second vs Not Addressed across 6 models
import numpy as np
import matplotlib.pyplot as plt

# ---- EDIT THESE VALUES ----
models = ["ChatGPT-4.1","ChatGPT-4o", "Flash 2.5","Flash-Lite 2.5", "LLaMA-4 Maverick","LLaMA-4 Scout"]
original = [136, 198, 330, 301, 201, 66]     # Excel Stats
epp      = [305, 241, 97, 16, 227, 254]      # Excel Stats
not_addressed = [0, 2, 14, 124, 13, 121]     # New Category
# ---------------------------

if len(models) != 6:
    raise ValueError("Please provide exactly 6 model names.")
if not (len(original) == len(epp) == len(not_addressed) == len(models)):
    raise ValueError("Each data list must have exactly 6 numbers.")

x = np.arange(len(models))
width = 0.25  # Slightly narrower to fit 3 bars per group

fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - width, original, width, label="Original Prompt")
bars2 = ax.bar(x, epp, width, label="EPP")
bars3 = ax.bar(x + width, not_addressed, width, label="Not Addressed")

ax.set_title("Analysis of Original Prompt vs EPP vs Not Addressed")
ax.set_ylabel("Weed Detected")
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=15, ha="right")
ax.legend()

# Annotate all bars with their values
for bars in (bars1, bars2, bars3):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f"{height:g}",
                    xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha="center", va="bottom")

plt.tight_layout()
plt.savefig("original_vs_epp_vs_notaddressed.png", dpi=200, bbox_inches="tight")
plt.show()
```

```python
# Original vs Second vs Not Addressed across 6 models
import numpy as np
import matplotlib.pyplot as plt

# ---- EDIT THESE VALUES ----
models = ["ChatGPT-4.1", "ChatGPT-4o", "Flash 2.5", "Flash-Lite 2.5", "LLaMA-4 Maverick", "LLaMA-4 Scout"]
original = [136, 198, 330, 301, 201, 66]      # Excel Stats
epp = [305, 241, 97, 16, 227, 254]            # Excel Stats
not_addressed = [0, 2, 14, 124, 13, 121]      # New Category
# ---------------------------

if len(models) != 6:
    raise ValueError("Please provide exactly 6 model names.")
if not (len(original) == len(epp) == len(not_addressed) == len(models)):
    raise ValueError("Each data list must have exactly 6 numbers.")

x = np.arange(len(models))
width = 0.25  # Narrower bars to fit three per group

fig, ax = plt.subplots(figsize=(11, 6))  # Slightly wider figure

# Create bars
bars1 = ax.bar(x - width, original, width, label="Original Prompt")
bars2 = ax.bar(x, epp, width, label="EPP")
bars3 = ax.bar(x + width, not_addressed, width, label="Missed Detections")

# Labels and title
ax.set_title("Analysis of Original Prompt vs Error Probing Prompting (EPP) vs Missed Detections")
ax.set_ylabel("Weed Detected")
ax.set_xticks(x)
ax.set_xticklabels(models)  # Keep them horizontal
ax.legend()

# Annotate all bars with values
for bars in (bars1, bars2, bars3):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f"{height:g}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha="center", va="bottom")

plt.tight_layout(rect=[0, 0.02, 1, 1])  # Add a little extra bottom space
plt.savefig("original_vs_epp_vs_notaddressed.png", dpi=200, bbox_inches="tight")
plt.show()
```
